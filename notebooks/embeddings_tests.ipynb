```python
import psycopg2, json
import google.generativeai as genai
```

```python
conn = psycopg2.connect("dbname=emb user=vagner")
cur = conn.cursor()
```

```python
#cur.execute("CREATE TABLE bcorpus (id bigserial PRIMARY KEY, doc INT, page INT, txt VARCHAR(10000), embedding vector(768));")
```

```python
api_key = open('../auth/gemini_key.txt').read()
genai.configure(api_key=api_key)
```

```python
def create_embedding(txt):
    result = genai.embed_content(
        model="models/text-embedding-004",
        content=txt)
    
    return result['embedding']
```

```python
query = 'What is the maximum rating for RCBO in the small power system?'
emb = create_embedding(query)
```

```python
db_query = f"SELECT * FROM bcorpus ORDER BY embedding <-> '{emb}' LIMIT 5;"
cur.execute(db_query)
```

```python
res = cur.fetchall()
```
